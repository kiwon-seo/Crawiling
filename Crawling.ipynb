# Oscar 수상자 크롤링 Tutorial

```python
import time
import pandas as pd

from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome()  # Optional argument, if not specified will search path.
driver.get('https://oscar.go.com/winners');
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

category = soup.select(
    'bls-winners-list > ul > li > div.winners-list__info > a'
)
movie = soup.select(
    'bls-winners-list > ul > li > div.winners-list__info > h3 > a'
)
producer = soup.select(
    'bls-winners-list > ul > li > div.winners-list__info > p'
)
oscars_2020 = []
for item in zip(category, movie, producer):
    oscars_2020.append(
        {
            'category' : item[0].text,
            'movie'    : item[1].text,
            'producer' : item[2].text
        }
    )
```

# 네이버스 리뷰 크롤링 by Selenium

```python
import time
import pandas as pd
import re

from selenium import webdriver
from bs4 import BeautifulSoup

# webdriver 실행. 같은 폴더 내에 webdriver.exe 파일이 있으므로 따로 파일 경로 지정할 필요 없음.
driver = webdriver.Chrome()  

# driver가 지정 url을 가져온다.
driver.get('https://play.google.com/store/apps/details?id=com.neibus&showAllReviews=true');

#브라우저에서 사용되는 엔진 자체에서 파싱되는 시간을 기다려 주는 메소드로 5초동안 wait하게 함.
driver.implicitly_wait(5)



## 스크롤 다운 해야 데이터가 생기는 동적 수집을 해야하는 경우이므로 아래와 같은 코드를 사용

SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom       0 ~ last_height까지 script를 실행시킴
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page         -50을 하는 이유는 script를 읽을 때 추가적인 loading을 피하기 위해.
    time.sleep(SCROLL_PAUSE_TIME)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height      새롭게 loading 된 script의 heigth를 new_heigth로 지정.
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    
    #  그렇게 생긴 new_height가 더이상 늘어나지 않는다면 break.
    if new_height == last_height:
        break

    last_height = new_height

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

review=str(soup.find_all('span', jsname='bN97Pc'))
review = re.sub('<.+?>', '', review, 0).strip()
review
```

위와 같이 '네이버스'의 모든 리뷰를 크롤링하여 정제했다. 이제 이것을 통해 워드클라우드 작업을 해보자.

### WordCloud 

```python
from collections import Counter
from konlpy.tag import Twitter
import pytagcloud
```

```python
a=str(review)
```

```python
nlp = Twitter()  
nouns = nlp.nouns(a)
```

```python
count = Counter(nouns)
count
```

```python
count.most_common(10)
```

```python
tag = count.most_common(40)
taglist = pytagcloud.make_tags(tag, maxsize=80)
```

```python
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900, 600), fontname='NanumBarunGothic', rectangular=False)
```

pytagcloud는 한글 폰트가 없기 때문에 따로 font를 복사해서 옮겨주고 fontname을 그에 따른 것으로 Wordcloud작업을 진행하며  
wordcloud.jpg라는 파일로 저장된다.
